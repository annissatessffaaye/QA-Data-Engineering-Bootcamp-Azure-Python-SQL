```python
import pandas as pd
import numpy as np
data = pd.read_csv('https://raw.githubusercontent.com/a-forty-two/cylons/master/data.csv', header=0)
data.tail(4)
```

```python
# pandas function-> .corr()
all_correlations = data.corr()
all_correlations
```

```python
# VISUALIZATION to understand correlation is a LOT easier than looking at the numbers!
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure()
sns.heatmap(all_correlations, cmap='coolwarm')
plt.show()
```
